```python
import numpy as np
import pandas as pd
import folium
import pandas as pd
import geopandas as gpd
import branca.colormap as cm
```

```python
data = pd.read_csv("https://covid-nma.com/dataviz/data/database_valid.csv", sep = ";",encoding='latin-1').drop(index = [0])
data = pd.DataFrame(data)
```

```python
# dimension of data
data.shape
```

```python
data_country = data[["Countries","Trial ID"]]
# change all the splitors to ; and delete space after , or ;
data_country['Countries'] = data_country.apply(lambda row: row['Countries'].replace(",",";")
                                               .replace(", ",";")
                                               .replace("; ",";"),axis = 1)
data_country['Countries'] = data_country.apply(lambda row: row['Countries'].replace("South Korea","Korea")
                                                                        .replace("Republic of Korea", "Korea")
                                                                        .replace("Korea","South Korea")
                                                                        .replace("United\xa0Kingdom", "United Kingdom")
                                                                        .replace("Hong Kong", "China")
                                                                        .replace('India \xa0',"India")
                                                                        .replace('South\xa0Africa', "South Africa")
                                                                        .replace('United States Minor Outlying Islands', "United States")
                                                                        .replace("Republic of the Congo", "Congo")
                                                                        .replace("India \xa0","India")
                                                                        ,axis = 1
                                              )
data_country["Countries"] = data_country.apply(lambda row: row["Countries"].split(";"), axis=1)
```

```python
data_grouped = data_country.loc[:,["Countries","Trial ID"]].groupby(["Trial ID"])
```

```python
all_countries = []
nb_trials = {}
for i in range(0,len(list(data_grouped))) :
    countries = list(data_grouped)[i][1].reset_index().loc[0,["Countries"]][0]
    nb = list(data_grouped)[i][1].shape[0]
    for country in countries :
        if country not in all_countries :
            all_countries.append(country)
            nb_trials[country] = 1
            print(country)
        else :
            nb_trials[country] = nb_trials[country] + 1
```

```python
# transform dictionary into dataframe
df_trials = pd.DataFrame(pd.Series(nb_trials),columns=["Nombre d'essais"])
df_trials = df_trials.reset_index().rename(columns = {"index" : "Country"})
```

```python
#define countries boundries using gpd
country_boundries = gpd.read_file('country_boundries/country_boundries.shp')
country_boundries = country_boundries[["CNTRY_NAME","geometry"]]
```

```python
country_boundries = country_boundries.rename(columns = {"CNTRY_NAME" : "Country"})
df_trials = pd.merge(df_trials, country_boundries, how='left', on='Country')
```
