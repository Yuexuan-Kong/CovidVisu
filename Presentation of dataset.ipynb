```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
# data = pd.read_csv(r'database_valid.csv', encoding = 'latin-1', sep = ';')
data = pd.read_csv("https://covid-nma.com/dataviz/data/database_valid.csv", sep = ";",encoding='latin-1').drop(index = [0])

df = pd.DataFrame(data)
# print(df)
```

```python
df["Number of arms"].describe()
```

```python
# Total trial numbers
nb_trial = len(df["Trial registration number"].unique())
print("Total trial numbers is",nb_trial)
```

```python
print(df.loc[:,'Treatment type'].value_counts().describe())
fig = plt.figure(figsize=(20,4))
df.loc[:,'Treatment type'].value_counts()[0:50].plot.bar(fontsize=15)
plt.savefig("treatment type",bbox_inches="tight")
```

```python
print(df.loc[:,'Treatment name'].value_counts().describe())
fig = plt.figure(figsize=(20,4))
df.loc[:,'Treatment name'].value_counts()[0:50].plot.bar(fontsize=15)
plt.savefig("treatment name",bbox_inches = "tight")
```
